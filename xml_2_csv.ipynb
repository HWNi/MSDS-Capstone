```python
data_list = list()
for file in os.listdir(path):
    if file.endswith("xml") and file in valid_list:
        print("Processing:", file)
        current_xml = os.path.join(path, file)
        parsed_xml = et.parse(current_xml)
        root = parsed_xml.getroot()
        #each_row = [None]*11
        for child in root:
            if child.tag == "personID":
                person_id = child.text
                #each_row[0] = child.text
            if child.tag == "FullName":
                full_name = child.text
                #each_row[1] = child.text
            if child.tag == "FirstName":
                first_name = child.text
                #each_row[2] = child.text
            if child.tag == "LastName":
                last_name = child.text
                #each_row[3] = child.text
            if child.tag == "publication":
                each_row = [None]*11
                for subchild in child:
                    if subchild.tag == "title":
                        each_row[4] = subchild.text
                    if subchild.tag == "year":
                        each_row[5] = subchild.text
                    if subchild.tag == "authors":
                        each_row[6] = subchild.text
                    if subchild.tag == "jconf":
                        each_row[7] = subchild.text
                    if subchild.tag == "id":
                        each_row[8] = subchild.text
                    if subchild.tag == "label":
                        each_row[9] = subchild.text
                    if subchild.tag == "organization":
                        each_row[10] = subchild.text
                each_row[0] = person_id
                each_row[1] = full_name
                each_row[2] = first_name
                each_row[3] = last_name
                data_list.append(each_row)
                #print(each_row)
```

### Process List

```python
# output_list contains single element
def process_list_v1(input_list):
    output_list = [None]*30000
    
    for index, value in input_list:
        output_list[index] = value
    
    return output_list


# output_list contains a list of elements
def process_list_v2(input_list):
    
    output_dict = dict()
    for index, value in input_list:
        if index in output_dict:
            output_dict[index].append(value)
        else:
            output_dict[index] = [value]

    output_list = [None]*30000
    
    for key, value in output_dict.items():
        output_list[key] = ", ".join(value)
            
    return output_list
```

### ArticleTitle

```python
titles_list = []
for i, element in enumerate(root.iter("MedlineCitation")):
    for child in element:
        if child.tag == "Article":
            for sub_child in child.iter('ArticleTitle'):
                titles_list.append((i, sub_child.text))
len(titles_list)
```

### PMID

```python
pmid_list = []
for element in root.iter('MedlineCitation'):
    pmid_list.append(element.find('PMID').text)
    
print(len(pmid_list))
```

### DateCompleted

```python
year_list = []
for i, element in enumerate(root.iter('MedlineCitation')):
    if element.find('DateCompleted')[0].tag == "Year":
        print(i, element.find('DateCompleted')[0].text)
        
len(year_list)
```

### Language

```python
lang_list = []
for i, element in enumerate(root.iter("MedlineCitation")):
    for child in element:
        if child.tag == "Article":
            for sub_child in child.iter('Language'):
                lang_list.append((i, sub_child.text))

len(lang_list)
```

### KeywordList

```python
keyword_list = []
for i, element in enumerate(root.iter("MedlineCitation")):
    for child in element:
        if child.tag == "KeywordList":
            for sub_child in child.iter('Keyword'):
                keyword_list.append((i, sub_child.text))
len(keyword_list)
```

### AbstractText

```python
abstract_text_list = []
for i, element in enumerate(root.iter("MedlineCitation")):
    for child in element:
        if child.tag == 'Article':
            for sub_child in child.iter('Abstract'):
                abstract_text_list.append((i, sub_child.find('AbstractText').text))

len(abstract_text_list)
```

```python
title = process_list_v1(titles_list)
```

```python
lang = process_list_v2(lang_list)
```

```python
keyword = process_list_v2(keyword_list)
```

```python
abstract_text = process_list_v1(abstract_text_list)
```

```python
combined_list = list(zip(pmid_list, title, lang, keyword, abstract_text))
```

```python
headers = ["PMID", "Title", "Language", "Keyword", "Abstract_Text"]
df = pd.DataFrame(combined_list, columns=headers)
df.to_csv('paper_big.csv', index=False)
```

### Parsing Name -> author_paper.csv

```python
import xml.etree.cElementTree as et
import pandas as pd

parsed_xml = et.parse("pubmedsample18n0001.xml")
root = parsed_xml.getroot()

big_author_list = []

def generate_author_paper(root):
    for i, element in enumerate(root.iter("MedlineCitation")):
        for child in element:
            if child.tag == 'Article':
                for sub_child in child.iter("Author"):
                    author = [element.find('PMID').text, None, None, None, None, None]
                    for ele in sub_child.getchildren():
                        if ele.tag == 'LastName':
                            author[1] = ele.text
                        elif ele.tag == 'ForeName':
                            author[2] = ele.text
                        elif ele.tag == 'Initials':
                            author[3] = ele.text
                        elif ele.tag == "AffiliationInfo":
                            author[4] = ele.find('Affiliation').text
                        elif ele.tag == "Identifier":
                            author[5] = ele.text
                    big_author_list.append(author)
    headers = ["PMID", "LastName", "ForeName", "Initials", "AffiliationInfo", "Identifier"]
    df = pd.DataFrame(big_author_list, columns=headers)
    df.to_csv('author_paper_big.csv', index=False)
```
